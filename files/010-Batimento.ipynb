# Batimento

Vamos considerar duas funções harmônicas, com mesma amplitude e frequências levemente diferentes,
$$x_1(t) = X \cos(\omega t); \quad
x_2(t) = X \cos\left((1+\delta)\omega t\right).
$$

A constante $\delta$ é a variação percentual da frequência, e, para configurar o batimento, dever ser pequena, da ordem de 0.01. 

Atenção: No livro, $\delta$ é uma variação absoluta! Usamos uma variação relativa aqui pois é mais fácil de programar.

## Setup

```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

## Cálculo das funções e da soma

```python
ω = 440.0     # base frequency
δ = 0.04    # Frequency variation

T = 10      # Playing time, seconds
sr = 22050  # Sampling rate
t = np.linspace(0, T, int(T*sr), endpoint=False) # time variable
Tp = t.size // 5  # Plot time

x1 = np.cos(2*np.pi*ω*t)
x2 = np.cos((2*np.pi*1.0+δ)*ω*t)
x = x1+x2

fig, ax = plt.subplots()
fig.set_size_inches((10,4))
ax.set_title('Sum of cosines')
ax.plot(t[:Tp], x1[:Tp], alpha=0.3)
ax.plot(t[:Tp], x2[:Tp], alpha=0.3)
_ = ax.plot(t[:Tp], x[:Tp])
```

## Play audio

```python
ipd.Audio(x, rate=sr) # load a NumPy array
```
