# Decremento logarítmico

Vamos ver o que acontece com a amplitude entre dois ciclos consecutivos para diferentes valores da razão de amortecimento $\zeta$, para ter uma noção qualitativa do efeito do amortecimento.

```python
# Setup
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Sabemos que 
$$\delta = \ln \left(\frac{x_1}{x_{2}} \right) = \zeta \omega_n \tau_d,$$
ou
$$ \ln \frac{x_1}{x_{2}} = \frac{2 \pi \zeta}{ \sqrt{1-\zeta^2}},$$
então
$$ \frac{x_1}{x_{2}} = e^{ \frac{2 \pi \zeta}{ \sqrt{1-\zeta^2}}}.$$


Para facilitar a interpretação, vamos visualizar $x_2/x_1$.


```python
# Make plot
zmax = 0.5
zetas = np.linspace(0, zmax, 201, endpoint=True) 
ratio = np.exp(2*np.pi*zetas/(np.sqrt(1.0-zetas**2)))
ratio = 1.0/ratio
fig, ax = plt.subplots()
fig.set_size_inches(14, 8)
ax.plot(zetas, ratio)
ax.set_xlim(left=0)
ax.set_ylim(bottom=0)
ax.set_title(r"Razão entre amplitudes consecutivas", fontsize=22)
ax.set_xlabel(r"$\zeta$", fontsize=18)
ax.set_ylabel(r"${x_2}/{x_1}$", fontsize=18)
ax.grid()
fig.show()



```

Também é interessante verificar o número de iterações necessário, por exemplo, para a amplitude decrescer a um valor pequeno, por exemplo, 1%, da amplitude inicial.

Neste caso usaremos a definição do decremento logarítmico ampliada para $m$ períodos,
$$ \delta = \frac{1}{m}\ln \left(\frac{x_1}{x_{m+1}}\right) = \frac{2 \pi \zeta}{ \sqrt{1-\zeta^2}},$$
ou
$$ m = \ln \left(\frac{x_1}{x_{m+1}}\right) \frac{\sqrt{1-\zeta^2}}{2 \pi \zeta} . $$

Vamos chamar a razão $x_1/x_{m+1}$ de $r_m$ no código abaixo.

```python
rm = 300
zmin = 1e-3
zmax = 0.5
zetas = np.geomspace(zmin, zmax, 501, endpoint=True)
m = np.log(rm)*np.sqrt(1-zetas**2)/(2*np.pi*zetas)
fig, ax = plt.subplots()
fig.set_size_inches(14, 8)
ax.plot(zetas, m, label=f"rm = {rm}")
ax.set_xlim(left=zmin)
ax.set_ylim(bottom=1)
ax.set_title(r"Número de iterações para $x_1/x_{m+1} = r_m$", fontsize=22)
ax.set_xlabel(r"$\zeta$", fontsize=18)
ax.set_ylabel("m", fontsize=18)
ax.set_yscale("log")
ax.set_xscale("log")
ax.grid(which="both")
ax.legend(fontsize=20)
fig.show()
```
